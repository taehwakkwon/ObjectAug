```python
import numpy as np

import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from torch.utils.data import Dataset, DataLoader
import cv2

colors = [0,(0,0,0),(255,255,255), (255,0,0),(0,255,0),(0,0,255),(255,255,0),(0,255,255), (255,0,255), (192,192,192), (128,128,128), (128,0,0)]


labels = ("UNKNOWN", "General trash", "Paper", "Paper pack", "Metal", "Glass", 
           "Plastic", "Styrofoam", "Plastic bag", "Battery", "Clothing")
```

```python
from tqdm import tqdm

def _draw_rectangle(image, label, score, y, x, height, width):
    height -= y
    width -= x
    cv2.rectangle(image, (y, x), (y+height, x+width), colors[label], 2)
    cv2.putText(image, f'{labels[label]}{score}', (y, x+width//2), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, colors[label], 2)
    return image

def draw_rectangle(image, target):
    # bboxes = sorted(bboxes, key=lambda x:x[2], reverse=True)[:10]
    for idx, box in enumerate(target['boxes']):
        label = target['labels'][idx]
        score = 0
        y, x, height, width = map(int, box)
        image = _draw_rectangle(image, int(label.detach().cpu().numpy()), score, y, x, height, width)
    
    return image

def gen_img(dataset, choices, base_dir='data'):
    n = len(choices)
    fig = plt.figure(figsize=(20, 200))
    
    for idx, index in enumerate(choices, start=1):
        image, target, image_id = dataset[index]
        image = image.permute(1,2,0).numpy()

        ax = fig.add_subplot(n,3,idx)
        ax.imshow(draw_rectangle(image, target))
        ax.set_title(f'{image_id} : original')

    plt.show()
        
n = len(dataset)

choices = np.random.randint(0, n, size=30)
gen_img(dataset, choices)
```
